```python
from pypdf import PdfReader

def print_pdf_content_as_text(pdf_path):
    # Open and read the PDF file
    reader = PdfReader(pdf_path)
    
    # Iterate over each page in the PDF and extract text
    for page_num, page in enumerate(reader.pages, start=1):
        print(f"--- Page {page_num} ---")
        text = page.extract_text()
        if text:
            print(text)
        else:
            print("[No text extracted from this page]")
            
if __name__ == '__main__':
    # Replace 'example.pdf' with your PDF file path
    pdf_file = "/Users/ashmitadua/Documents/VSCodeProjects/Hackathons/CruzHacks/CSSlugHandbook/SlugBot/SSR_TSRPT (3).pdf"
    print_pdf_content_as_text(pdf_file)
```
